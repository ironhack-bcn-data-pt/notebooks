## Recursive functions

```python
def factorial(num):
    fac = 1
    for i in range(1,num+1):
        fac *= i
    return fac
```

```python
factorial(8)
```

```python
def factorial(num):
    if num==1:
        return 1
    else:
        return num * factorial(num-1)
```

```python
factorial(5)
```

```python
def factorial(num):
    if num==1:
        return 1
    
    return num * factorial(num-1)
```

```python
from math import sqrt

def is_prime(num):
    isprime = True

    for x in range(2, int(sqrt(num) + 1)):
        if num % x == 0: 
            isprime = False
            break

    return isprime
```

```python
is_prime(109)
```

```python
def is_prime(num):

    for x in range(2, num):
        if num % x == 0: 
            return False

    return True
```

```python
is_prime(15)
```

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```

```python
def genera_primos():
    num = 1
    while True:
        if is_prime(num):
            yield num
        num += 1
```

```python
for i in genera_primos():
    print(i)
```

## zip

```python
name = ['John', 'Danny', 'Tyrion', 'Sam']
depto = ['Sales', 'Marketing', 'IT', 'Human Resource']
sal = [1000,1500,1200,2000]
for n,d,s in zip(name, depto, sal):
    print(f'{n} works at {d} and his/her salary is {s}')
```

## Decorators

```python
def do_twice(func):
    
    def wrapper():
        func()
        func()
        
    return wrapper
```

```python
hola = do_twice(hola)
```

```python
@do_twice
def hola():
    print('Hola mundo')
```

```python
hola()
```
