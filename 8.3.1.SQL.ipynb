## Exercises

- Número de tracks por album
- Album con más tracks
- Duración media de las canciones de cada album
- Las 10 canciones más largas

```python
%%sql
select album, count(id)
from Track 
group by album
```

```python
%%sql
select album, count(id) as num_tracks
from Track 
group by album
order by num_tracks desc
limit 1
```

```python
%%sql
select album, round(avg(duration)) as mean_duration
from Track 
group by album
order by mean_duration desc
limit 10 
```

```python
%%sql
select distinct name, duration
from Track 
order by duration desc
limit 10 
```

```python
%%sql
select a.name as album, t.name as camcion, t.duration as duracion, a.year as year
from Track t inner join Album a on t.album = a.id
limit 20
```

```python
%%sql
select a.name as album, round(avg(t.duration)) as duracion_media
from Track t inner join Album a on t.album = a.id
group by a.name
```

```python
%%sql
select ar.name as artist, a.name as album, round(avg(t.duration)) as duracion_media
from Track t inner join Album a on t.album = a.id
             inner join Artist ar on a.artist = ar.id
group by ar.name, a.name
```

```python
query = '''
select ar.name as artist, a.name as album, round(avg(t.duration)) as duracion_media
from Track t inner join Album a on t.album = a.id
             inner join Artist ar on a.artist = ar.id
group by ar.name, a.name
'''
```
