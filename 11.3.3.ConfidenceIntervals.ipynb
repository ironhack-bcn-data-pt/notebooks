```python
import scipy.stats as stats
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
samples = [np.mean(stats.bernoulli(0.65).rvs(1000)) for i in range(1000)]
```

```python
sns.distplot(samples, bins=10)
plt.vlines(np.mean(samples), 0, 30)
```

```python
confidence = 0.95

n = len(samples)
m = np.mean(samples)
std_err = stats.sem(samples)
std = np.std(samples)

h = std * stats.t(n -1).ppf((1 + confidence) / 2)

start = m - h
end = m + h

print('Confidence interval ({:.0f}%): {} - {}'.format(confidence*100, start,end))
```

```python
sns.distplot(samples, bins=10)
plt.vlines([start, end], 0, 30, color='red')
```
