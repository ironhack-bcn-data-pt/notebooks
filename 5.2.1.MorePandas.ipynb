# Reshape

```python
df = pd.read_csv('data/life_expectancy_years.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df_long = pd.melt(df, id_vars=['country'], var_name='year', value_name='life_expectancy')
```

```python
df_long.head()
```

```python
df_long.country.value_counts()
```

```python
df_long['life_expectancy'] = df_long['life_expectancy'].str.replace(',','.').astype('float')
```

```python
pd.pivot_table(df_long, index='country', values='life_expectancy', columns='year')
```

## Concat

```python
a = np.random.randint(0,9,(10,4))
a = pd.DataFrame(a, columns=['a','b','c','d'])
a
```

```python
b = np.random.randint(0,9,(10,5))
b = pd.DataFrame(b, columns=['a','b','c','e','f'])
b
```

```python
pd.concat([a,b], join='outer')
```

```python
pd.concat([a,b],axis=1)
```

## Merge

```python
 

a = pd.DataFrame({'a':['a','b','c'], 'b':[1,2,3]})
a
```

```python
b = pd.DataFrame({'a':['c','b','d'], 'b':[10,20,30]})
b
```

```python
pd.concat([a,b], axis=1)
```

```python
pd.merge(a, b, on='a', how='inner')
```
