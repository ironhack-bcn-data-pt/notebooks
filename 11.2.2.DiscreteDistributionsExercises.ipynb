## Die sum distribution

Given X = {Sum of numbers on the die when rolled twice}, plot the PMF of P(X)

```python
np.random.randint(1,7,n)
```

```python
list(zip(np.random.randint(1,7,n), np.random.randint(1,7,n)))
```

```python
[a + b for (a,b) in zip(np.random.randint(1,7,n), np.random.randint(1,7,n))]
```

```python
n = 10000
h = plt.hist([a + b for (a,b) in zip(np.random.randint(1,7,n), np.random.randint(1,7,n))], bins=11)
```

```python
h[0]
```

```python
h[0]/n
```

## PMF

```python
plt.bar(x=range(2,13), height=(h[0]/n))
```

## CDF

```python
plt.bar(x=range(2,13), height=np.cumsum(h[0]/n))
```

```python
np.cumsum(h[0]/n)
```

## SF (Survival function)

```python
1 - np.cumsum(h[0]/n)
```

```python
plt.bar(x=range(2,13), height=1-np.cumsum(h[0]/n))
```
